```python
#Dominios lineales
x = arange(0,1,1/512)
y = arange(0,1,1/512)

xx, yy = meshgrid(x,y)
```

```python
imshow(xx)
```

```python
imshow(xx+yy)
```

```python
imshow(xx**2+yy**2)
```

```python
zz = (xx-0.5)**2+(yy-0.5)**2
imshow(zz)
```

```python
imshow(exp(-zz)) #Gaussiana
```

```python
zz = exp(-zz)
imshow(zz)
```

```python
F2zz = fft.fft2(zz)
F2zz.shape
```

```python
imshow(abs(F2zz))
```

```python
imshow(20*log10(abs(F2zz)))
```

```python
imshow(20*log10(abs(F2zz)))
xlim(0,12)
ylim(0,12)
```

```python
zz = cos(2*pi*100*(xx+0.5*yy))
imshow(zz)
```

```python
F2zz = fft.fft2(zz)
imshow(20*log10(abs(F2zz)))
```

```python
imshow(20*log10(abs(F2zz)))
xlim(90,110)
ylim(40,60)
```
