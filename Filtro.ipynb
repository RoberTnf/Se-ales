```python
%pylab inline
import simdaq as s
rcParams['figure.figsize'] = (10.0, 9.0)
```

```python
sig1=s.Signal(t1=-5,t2=5,fm=8000)
sig1.pulse(pos=0,wid=1)
```

```python
sig1.pinta()
```

```python
sis = s.LTI_System(T=10,fm=8000)
sis.filtroPasivoPasaBajos(300,5e-2,1e-8)
sig2 = sis(sig1)
sig2.pinta(xlim=(-1,1))
```

```python
sig1=s.Signal(t1=-5,t2=5,fm=8000)
sig1.tone(fr=10)
sig1.pinta()
```

```python
sis = s.LTI_System(T=10,fm=8000)
sis.filtroPasivoPasaBajos(R=2000,L=5e-2,C=1e-11)
sig2 = sis(sig1)
sig2.pinta()
```
